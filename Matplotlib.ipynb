```python
#simple plot

from matplotlib import pyplot as plt
import numpy as np
import math

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)      

plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("ohh")
plt.show()
```

```python
#histogram

from matplotlib import pyplot as plt
import numpy as np

fig,ax=plt.subplots(1,1)
a=np.array([22,45,67,5,6,78,89,56,45,67,21,5,8,56,98,8])
ax.hist(a,bins=[0,25,50,75,100])
ax.set_title("he")

ax.set_xticks([0,25,50,75,100])
ax.set_xlabel("training")
ax.set_ylabel("no of students")

plt.show()
```

```python
#pie chart

from matplotlib import pyplot as plt
import numpy as np

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')

course=["aws","bigdata","python","datascience"]
student=[23,23,7,89]
ax.pie(student, labels=course,autopct='%9.3f')
plt.show()
```

```python
#pie chart for coronavirus cases

from matplotlib import pyplot as plt
import numpy as np

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')

course=["karnataka","Tamil Nadu","Telangana","Kerela"]
student=[23,23,7,89]
ax.pie(student, labels=course,autopct='%9.3f%%')
plt.show()
ax.set_xlabel("coronoavirus in south")
```

```python
#bar chart for coronavirus cases

from matplotlib import pyplot as plt

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
course=["karnataka","Tamil Nadu","Telangana","Kerela"]
student=[23,23,87,89]
ax.bar(course,student)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("coronoavirus in south")
plt.show()
```

```python
#scatter plot

import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
z = [5,7,8,7,2,17,2,9,4,11,12,9,6]

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='b')
ax.scatter(y,z,color='g')

ax.set_xlabel('x')
ax.set_xlabel('y')
ax.set_xlabel('z')

#plt.scatter(x, y)
plt.show()
```
