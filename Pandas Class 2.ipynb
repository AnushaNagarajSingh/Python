```python
#concatenate

import pandas as pd
India_weather=pd.DataFrame({
    "city":["mumbai","Delhi","Bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
India_weather
```

```python
#concatenate

import pandas as pd
us_weather=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
us_weather
```

```python
df=pd.concat([India_weather, us_weather],axis=1)
df
```

```python
df=pd.concat([India_weather, us_weather],axis=0)
df
```

```python
humidity_weather=pd.DataFrame({
    "city":["new york","chicago","orlando","dubai"],
    "humidity":[54,87,78,50]
})
humidity_weather
```

```python
temp_weather=pd.DataFrame({
    "city":["new york","chicago","orlando","dubai"],
    "humidity":[54,87,78,50]
})
temp_weather
```

```python
#merge without explicitly mention

df=pd.merge(temp_weather,humidity_weather,on="city")
df
```

```python
#outer join

df=pd.merge(temp_weather,humidity_weather,on="city",how="outer")
df
```
